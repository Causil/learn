    ¡Bienvenido al curso DevNet Associate! Este curso le enseñará los conceptos básicos del desarrollo de software, los fundamentos de redes y la automatización.

La industria está cambiando hacia la automatización, las aplicaciones y las integraciones. Es importante aprender las habilidades necesarias para mantenerse al día con el paso acelerado de la tecnología. Los módulos de este curso se eligieron específicamente para ayudarlo a obtener habilidades laborales críticas.

La causa de este cambio en la industria son las API (Application Programing Interfaces). Necesitamos ser más eficientes en la forma en que implementamos dispositivos, aplicaciones, redes e infraestructuras. Una forma de hacerlo es eliminar las tareas manuales repetitivas con la automatización computarizada mediante API. Las API permiten crear scripts automatizados, así como integraciones que construyen una red entre aplicaciones. Esto permite que las aplicaciones compartan datos, soliciten la realización de algunas acciones o incluso se notifiquen entre sí cuando ocurre un evento específico. Aprenderá acerca de las API, sus beneficios y cómo solucionar problemas en el módulo de uso y comprensión de las API.

Para poder utilizar las API para escribir scripts y crear aplicaciones, se debe saber cómo codificar. Debe aprender los principales conceptos de desarrollo de software y estar equipado con las herramientas necesarias para escribir código de calidad. Estos temas se tratarán en el módulo de Desarrollo y Diseño de Software.

La automatización y las integraciones mediante las API no puede existir sin una interconexión de red. La red permite la comunicación entre aplicaciones y clientes. El uso de las API ha ayudado a que la red sea más eficiente a través de la programabilidad de la red. La programabilidad de red es la capacidad de configurar, supervisar y reaccionar ante eventos en la red en tiempo real. Es importante aprender acerca de los conceptos básicos de la red, los cuales se discuten en detalle en el módulo Fundamentos de Red.

Con el creciente número de las API en la industria junto a la intergración de aplicaciones, es importante comprender cómo proteger la red y sus aplicaciones para evitar incidentes como las violaciones de datos. Existen diferentes problemas de seguridad para cada tipo de implementación, por lo que debe conocer los distintos modos de implementación para tomar la mejor opción para su aplicación. Más detalles sobre la protección de la red y sus aplicaciones se describen en el módulo Implementación y seguridad de aplicaciones.

Las API ya han abierto las puertas para la automatización de la red. La automatización de redes puede llevarse un paso más allá, mediante la implementación automatizada de aplicaciones. Puede hacer esto con la infraestructura como código. Esto es básicamente la capacidad de administrar la infraestructura con automatización, en lugar de configurar manualmente dicha infraestructura. Aprenderá más sobre este tema en el módulo de Infraestructura y Automatización.

El último Módulo es Plataformas y Desarrollo de Cisco. Este módulo lo reúne todo en grandes categorías, como la colaboración, la computación, incluyendo los centros de datos, además de las redes, incluyendo los modelos de datos, y la seguridad. ¡Comencemos!

<h1> Configure su entorno de laboratorio </h1>
<p>
Su entorno de laboratorio para este curso es en realidad un conjunto de máquinas virtuales. Con la virtualización, una o más computadoras virtuales funcionan en el interior de una computadora física. Las computadoras virtuales que funcionan en el interior de equipos físicos se llaman máquinas virtuales (Virtual Machines). A menudo, las máquinas virtuales se denominan guests (Huesped o invitado) y las computadoras físicas suelen llamarse anfitriones o hospedadores (hosts). Cualquier persona con una computadora y un sistema operativo moderno puede ejecutar máquinas virtuales.
</p>
