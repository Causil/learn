<h2> Objetos reales de lo que existe si logramos la meta </h2>
<h3> Tipos de páginas web </h3>
<ol>
    <li>
    Página de negocio:  Los sitios web para negocios suelen ser una recopilación de varias páginas que brindan información sobre la empresa y presentan sus productos o servicios.  
    </li>
    <li>
        Página web de blog: Los blogs son un tipo de página web que muestran extensos textos editoriales sobre prácticamente cualquier tema. Se componen principalmente de artículos e imágenes, normalmente en torno a un tema específico. Algunos de los tipos de blogs más populares son los blogs de viajes, blogs de comida y blogs de belleza.
    </li>
    <li>
     Página web de eCommerce o tienda online: 
        Similar a las páginas web de negocios mencionados anteriormente, los sitios web de eCommerce o tienda online, te permiten vender productos online. Sin embargo, para una amplia gama de productos, te recomendamos crear una tienda online.
    </li>
    <li>
    Página web de portafolio : A menudo, la gente utiliza páginas web para mostrar su propio trabajo o sus habilidades. Los portafolios son el tipo de página web más adecuado para este fin. Suelen contener una galería que muestra los mejores trabajos de una manera atractiva. El diseño de las páginas web de portafolio generalmente se adaptan al trabajo y la creatividad de cada profesional y, por lo tanto, refleja su estilo individual.
    </li>
    <li>
    Página web de evento: A medida que la gente se ha ido trasladando al mundo online, las páginas web de eventos se han hecho extremadamente populares. No importa qué tipo de evento estés planeando, un sitio web de eventos te ayudará a resumir toda la información importante sobre tu evento en un solo lugar, como horarios, sugerencias de transporte, mapas y mucho más. También pueden incluir un software de reservas y venta de entradas que facilita a la gente confirmar su asistencia. Esto es especialmente útil para conciertos, espectáculos o conferencias, y también puede utilizarse para eventos virtuales. 
    </li>
    <li>
        Página web de online: Otro producto de nuestra era virtual son los foros en línea. Los foros permiten a la gente discutir temas, hacer preguntas e interactuar entre sí. Cada vez son más las páginas web que incluyen un foro en línea para que sus miembros puedan hablar entre sí y fomentar la conversación en su sitio. Esto se usa a menudo como un tipo de soporte informal que permite a los usuarios del sitio ayudarse mutuamente. 
    </li>
    <li>
        Página web personal:Crear una página web personal es una forma sencilla de potenciar tu presencia en la red. Este tipo de sitio web es esencialmente un espacio que rodea y representa a una persona determinada. Muchas personas utilizan páginas web personales para mostrar quiénes son y lo que han logrado. En muchos sentidos es similar a un portafolio con menos elementos visuales, lo que lo convierte en el formato perfecto para sitios web de autores, actores, entre otros
    </li>
    <li>
        Página web de landing page: Una landing page o página de destino es una página web diseñada para lograr un solo objetivo: convertir visitantes en clientes. Se utilizan en estrategias de marketing y para campañas publicitarias. Este tipo de páginas aparecen después de hacer clic en Google, una campaña de marketing por correo electrónico, una campaña en redes sociales o un anuncio online. 
    </li>
    <li>
        Página web con membresías: Un método utilizado por muchos propietarios de páginas web es solicitar una membresía para proporcionar acceso completo a sus servicios, herramientas o recursos. Este tipo de página web suele permitir a los propietarios del sitio ganar dinero con sus ofertas a los miembros, construir una comunidad online y simplemente tener más control sobre la audiencia de su sitio. 
    </li>
    <li>Páginas web sin ánimo de lucro: Otro tipo de página web son aquellas para organizaciones sin ánimo de lucro. Al igual que los sitios web de negocios, estas páginas suelen incluir una descripción de la organización, detalles sobre el trabajo que realizan, una galería de eventos pasados y un espacio para hacer donaciones. </li>
    <li>Página web informativa: Las páginas web informativas son un tipo de sitio web que puede cubrir muchos temas diferentes. En general, se utilizan como plataformas cuyo objetivo principal es transmitir información y actuar como recurso. Los sitios web informativos no pretenden ser una fuente de ingresos, sino que simplemente pretenden informar.
</li>
    <li>Página web de multimedia: Las páginas web de multimedia son otro tipo de página web que te permite llevar tu trabajo al mundo online. Las páginas web de multimedia suelen mostrar videos, sonidos y contenido animado y dinámico en la página. Son una buena opción para quienes buscan promocionar, vender y distribuir su contenido multimedia desde una sola plataforma. 

  </li>
</ol>
</div>
