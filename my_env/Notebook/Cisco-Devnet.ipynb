<h3> Understand how devices find each other </h3>
<p> This Lab describes the processes behind the scenarios in LAN and Office network. </p>

<h4> Revisiting the LAN scenario: routers and switches </h4>
<p> 
In the LAN scenario, the network completes the following processes when a student sends a message to the instructor.
</p>
<h3> Simple Local Area Network (LAN) </h3>

<img src='https://developer.cisco.com/media/networking-101-the-basics/labs/assets/images/the-network.png'>

<p>
The student computer must know the IP address of the instructor's computer. In this example, the IPv4 address of the instructor's computer is 172.16.16.1 and the address of the student computer is 172.16.254.1. Because the instructor's computer is on another subnet, the student computer sends the message request to its default gateway IP, belonging to the router. The network switch and router complete the following processes to send the data. 
</p>

<ol>
    <li>
        Process One: MAC Address table: The switch receives the message request and looks in its MAC Address Table for the MAC address associated with the router IP. If the switch locates the MAC address, it sends the student message to the port connected to the router for further processing. If the switch does not have the MAC address in its cache, it floods the request to all ports. 
    </li>
        <li>
Process Two: ARP Cache: The router receives the message request and looks in its ARP cache for the MAC address that is associated with IP 172.16.16.1. If the router locates the MAC address and other identifying information, it sends the student message to the instructor computer. If the router does not have the MAC address in its cache, it initiates the next process.
    </li>
    <li>
Process Three: ARP request: The router uses a protocol that is called Address Resolution Protocol (ARP) to broadcast an ARP request to look for the IP address 172.16.16.1 on all network devices. If one of the devices has that IP address, the device sends an ARP response that identifies itself and includes the MAC address. The router then passes the message to that device. However, if no devices connected to that router have a matching IP address, the router drops the request. 
    </li>
    <li>
Process Four: MAC Address table The switch receives the message request from the router and looks in its MAC Address Table for the MAC address that is associated with the instructor's computer. If the switch locates the MAC address, it sends the student message to the port connected to the computer. If the switch does not have the MAC address in its cache, it floods the request to all ports. 
    </li>
</ol>
<h3>Revisiting the Office network scenario: switch or hub </h3>
<p> 
 The Office network consists of computers and a printer that are connected by a switch or a hub. This simple network does not include a router, but may connect the devices via a switch or a hub.
</p>
<h3> Office Network </h3>

<img src='https://developer.cisco.com/media/networking-101-the-basics/labs/assets/images/office-network.png'>

<p>
    In the switch scenario, a computer must be able to identify the IP address of the printer. Because there is no router in this network, the IPv4 addresses for all devices must be on the same network segment so that they can find each other. To be "on the same segment" means that an IPv4 address can be found without the need to route to another network segment.

In this example, the IPv4 address of the printer is 172.16.254.5 and the user's computer's IPv4 address is 172.16.254.1. When the user attempts to send data to the printer, the devices perform the following processes.
</p>


<ol>
<li>
    Process One: ARP cache: The user computer looks in its ARP cache for the MAC address that is associated with 172.16.254.5. If the computer locates the MAC address and other identifying information, it sends the data to the MAC address of the printer. If the computer does not have the MAC address in its table, it initiates the next process.
</li>
<li>
    Process Two: ARP request: The user computer sends an ARP Request to look for the IP address 172.16.254.5. The printer receives the request and sends an ARP Response that identifies itself and includes the MAC address. The computer sends the data to the MAC address of the printer.
</li>
<li>
    Process Three: MAC Address table: The switch receives the message request and looks in its MAC Address Table for the MAC address that is associated with the printer. If the switch locates the MAC address, it sends the message out the port that is connected to the printer. If the switch does not have the MAC address in its cache, it floods the request to all ports.
</li>
</ol>

<p>    
A hub, also known as a repeater, is less sophisticated than a switch. It sends, or repeats, all data that it receives to all devices that are connected to it.
</p>
<ol>
<li>
Process: Data repetition: The hub receives the data and sends it to every device that is connected to it. If one of the devices has the matching IP address, it responds.
</li>
</ol>
<p>
Hubs are less efficient than switches, because they do not have an ARP cache and cannot use ARP. Hubs always broadcast data to all ports, which can saturate the network, consume bandwidth, and make computers reject data.

Many switches have a feature that is called microsegmentation that provides the same bandwidth to all connected devices. However, with a hub, all connected devices must share a limited amount of bandwidth. With microsegmentation, network traffic flows smoothly without collisions.

Finally, switches often provide many configuration options, while hubs offer little or no configuration.
</p>

<h3>Fases del ciclo de vida del desarrollo de software</h3>

El ciclo de vida de desarrollo de software (SDLC) es el proceso de desarrollo de software que comienza a partir de una idea y termina con la entrega. Este proceso consta de seis fases. Cada fase toma la información de los resultados de la fase anterior. No hay SDLC estándar, por lo que las fases exactas pueden variar, pero las más comunes son:

Fase 1. Requerimientos y Análisis

Fase 2 Diseño

Fase 3. Implementación

Fase 4. Pruebas

Fase 5. Despliegue

Fase 6. Mantenimiento

Históricamente, los equipos de desarrollo por lo general siguen estas fases en el orden según el método de Cascada (Waterfall) El objetivo del método de Cascada (Waterfall) era completar cada fase de SDLC hasta el último detalle antes de pasar a la siguiente, nunca volver a una fase anterior, y escribir todo en el camino.

Aunque los métodos de Cascada (Waterfall) todavía se utilizan ampliamente hoy en día, gradualmente están siendo reemplazados por métodos más adaptables y flexibles que producen mejor software, más rápido y con menos complicación. Estos métodos se conocen colectivamente como "desarrollo ágil" (Agile)

Es importante entender que el SDLC se puede aplicar de muchas maneras diferentes. Sus fases se pueden repetir e invertir el orden. Se pueden realizar fases individuales en muchos niveles en paralelo (por ejemplo, los requisitos se pueden recopilar por separado para los detalles de la interfaz de usuario, las integraciones de back-end, los parámetros operativos y de rendimiento, etc.).

Vamos a ver las fases del SDLC con mayor detalle, y en su orden clásico (sólo recuerde: esta es una descripción, no una prescripción)

<h1> Fase de Requerimientos y Análisis </h1>

El objetivo de la fase de requisitos y análisis es responder a varios niveles de preguntas. Estos empiezan por explorar la situación actual de las partes interesadas, las necesidades y limitaciones, la infraestructura actual, etc. Determinando los problemas que este nuevo software necesita resolver.

Una vez que el problema está mejor definido, se pueden explorar cuestiones más concretas para determinar dónde y en qué contexto tendrá que vivir el nuevo software.

Cuando se compilan las respuestas a tales preguntas, es hora de comenzar a explorar requisitos más precisos centrándose en las funciones deseadas y la experiencia del usuario (UX).

Finalmente, el equipo comienza a evaluar las opciones de aqrquitectura para construir el software en sí. Para la mayoría de las aplicaciones empresariales, esto significa muchas iteraciones de definición de requisitos para el front-end y el back-end del software. También tendrá que proporcionar puntos de integración para otras aplicaciones, así como servicios para la administración del ciclo de vida.

Después de reunir los requisitos, el equipo analiza los resultados para determinar lo siguiente:

    ¿Es posible desarrollar el software de acuerdo con estos requisitos y se puede hacer según el presupuesto?
    ¿Existe algún riesgo para el cronograma de desarrollo y, de ser así, cuáles son?
    ¿Cómo se probará el software?
    ¿Cuándo y cómo se entregará el software?

Al final de esta fase, el método clásico de Waterfall (Cascada) sugiere la creación de un documento de especificación de requisitos de software (SRS) que establece dichos requisitos y el alcance del software, y lo confirma meticulosamente con las partes interesadas.

<h3> Fases de diseño e implementación </h3>

Diseño

La fase de diseño toma clásicamente el documento SRS de la fase de Requisitos y Análisis como entrada. Durante la fase de diseño, los arquitectos y desarrolladores de software diseñan el software basado en el SRS proporcionado.

Al final de la fase de diseño, el equipo crea documentos de diseño de alto nivel (HLD) y diseño de bajo nivel (LLD). El HLD ofrece una "visión de 10.000 pies" del software propuesto. Describe la arquitectura general, los componentes y sus relaciones, y puede proporcionar detalles adicionales. El LLD, basado en el documento HLD, describe con mucho más detalle la arquitectura de los componentes individuales, los protocolos utilizados para comunicarse entre ellos, y enumera las clases requeridas y otros aspectos del diseño.

Implementación

La fase de implementación toma clásicamente el HLD y el LLD de la fase de diseño como entrada.

La fase de implementación a menudo se denomina fase de codificación o desarrollo. Durante esta fase, los desarrolladores toman la documentación de diseño y desarrollan el código de acuerdo con ese diseño. Todos los componentes y módulos se construyen durante esta fase, lo que hace que la implementación sea la fase más larga del ciclo de vida. Durante esta fase, los ingenieros de pruebas también están escribiendo el plan de pruebas.

Al final de la fase de implementación, el código funcional que implementa todos los requisitos del cliente está listo para ser probado.

<h3>Fases de prueba, despliegue y mantenimiento.</h3>

Prueba

La fase de prueba clásicamente toma el código de software de la fase de implementación como entrada. Durante esta fase, los ingenieros de pruebas toman el código y lo instalan en el entorno de pruebas siguiendo lo estipulado en el plan de pruebas. El plan de prueba es un documento que incluye una lista de cada prueba a realizar para cubrir todas las características y funcionalidades del software, según lo especificado por los requisitos del cliente. Además de las pruebas funcionales, los ingenieros de pruebas también realizan:

    Pruebas de integración
    Pruebas de rendimiento
    Pruebas de seguridad

Cuando el código no pasa una prueba, el ingeniero de pruebas identifica el error, y éste se informa a los desarrolladores. Una vez solucionado el error, los ingenieros de prueba volverán a probar el software. Esto de ida y vuelta entre los ingenieros de prueba y desarrollo continúa hasta que todo el código pase todas las pruebas.

Al final de la fase de prueba, un software de alta calidad, libre de errores, está listo para la producción, en teoría. En la práctica, esto rara vez sucede. Los desarrolladores han aprendido cómo realizar pruebas de manera más eficiente, cómo incorporar pruebas en flujos de trabajo automatizados y cómo probar software en muchos niveles diferentes de detalle y abstracción: desde las más pequeñas definiciones de funciones de bajo nivel hasta agregaciones de componentes a gran escala. También han aprendido que el software nunca está libre de errores y, en cambio, debe hacerse observable, probado en producción y resistente para que pueda permanecer disponible y funcionar adecuadamente, a pesar de los problemas.

Implementación

La fase de implementación toma el software de la fase de prueba como entrada. Durante la fase de implementación, el software se instala en el entorno de producción. Si no hay problemas de implementación, el gerente de producto trabaja con los arquitectos e ingenieros calificados para decidir si el software está listo para ser lanzado.

Al final de la fase de implementación, el software final se libera a los clientes y otros usuarios finales.

Mantenimiento

Durante la fase de mantenimiento, el equipo:

    Proporciona soporte a los clientes
    Se encargada de la correción de errores encontrados en producción
    Trabajar en mejoras del software
    Reúne nuevas solicitudes del cliente

Al finalizar la fase de mantenimiento, el equipo está listo para trabajar en la siguiente iteración y versión del software, lo que devuelve el proceso al inicio del SDLC y a la fase de requisitos y análisis.

<h3>Metodologías de Desarrollo de Software</h3>

Una metodología de desarrollo de software también se conoce como modelo de ciclo de vida de desarrollo de software. Estas metodologías no son más que un conjunto de reglas, pasos, roles y principios. Existen muchas metodologías diferentes, pero nos centraremos en las tres más populares:

    Waterfall
    Agile
    Lean 

Cada metodología tiene sus pros y contras. Decidir qué usar depende de muchos factores, como el tipo de proyecto, la duración del proyecto y el tamaño del equipo de desarrollo.

El modelo de desarrollo en Cascada (Waterfall) es el modelo tradicional de desarrollo de software, y todavía se practica hasta el día de hoy. El modelo de desarrollo en Cascada (Waterfall ) es casi idéntico al ciclo de vida de desarrollo del software ya que cada fase depende de los resultados de la fase anterior.

En este modelo en Cascada (Waterfall), el agua fluye en una sola dirección. Con el método en Cascada (Waterfall) el proceso va en una dirección, y nunca puede ir hacia atrás. Piense en ello como una carrera de relevos en la que un corredor tiene que terminar su distancia antes de entregar el testigo a la siguiente persona, que lo está esperando. La batuta siempre se mueve hacia adelante.

Se dice que el modelo original de Waterfall (Cascada) fue creado por Winston W. Royce. Su modelo original constaba de siete fases:

    Requisitos del sistema
    Requisitos de software
    Análisis
    Diseño de programas
    Codificación
    Pruebas
    Operaciones

Como puede ver, el modelo de Waterfall (Cascada) es realmente sólo una iteración del ciclo de vida del desarrollo de software. Ahora hay muchas variaciones de las fases en el modelo de Waterfall (Cascada), pero la idea de que cada fase no puede solaparse y debe completarse antes de seguir adelante sigue siendo la misma.

Debido a que el resultado de cada fase es crítico para la siguiente, una decisión equivocada puede descarrilar toda la iteración; por lo tanto, la mayoría de las implementaciones del modelo en Cascada (Waterfall) requieren documentación que resuma los hallazgos de cada fase como entrada para la siguiente fase. Si los requisitos cambian durante la iteración actual, esos nuevos requisitos no se pueden incorporar hasta la siguiente iteración, lo que puede resultar costoso para grandes proyectos de software y causar retrasos significativos antes de que las funciones solicitadas estén disponibles para los usuarios.

El método Ágil es flexible y se centra en el cliente. Aunque ya se estaban practicando metodologías similares a Agile, este modelo no fue oficial hasta 2001, cuando diecisiete desarrolladores de software se unieron para encontrar una solución a sus frustraciones con las opciones del momento y elaboraron el Manifiesto para el Desarrollo de Software bajo la premisa de Agile, también conocido como el Manifiesto Agile.

Manifiesto Agile.

Según el Manifiesto Agile, sus valores de Agile son:

    Individuos e interacciones sobre procesos y herramientas
    Software funcional sobre documentación completa
    Colaboración con clientes sobre negociación de contratos
    Respuesta al cambio sobre el seguimiento de un plan

El manifiesto enumera doce principios diferentes:

    Orientado al cliente - Nuestra máxima prioridad es satisfacer al cliente a través de la entrega temprana y continua de software valioso.
    
    Acepte el cambio y adaptese - De la bienvenida a los requerimientos cambiantes, incluso en la etapa final del desarrollo Los procesos Agile aprovechan el cambio para la ventaja competitiva del cliente.
    Entrega frecuente de software funcional - Entregar software de trabajo con frecuencia, desde un par de semanas hasta un par de meses, con preferencia a la escala de tiempo más corta.
    Colaboración - Los empresarios y desarrolladores deben trabajar juntos diariamente durante todo el proyecto.
    Equipos motivados - Construir proyectos alrededor de individuos motivados. Bríndeles el entorno y el apoyo que necesitan, y confíe en ellos para hacer el trabajo.
    Conversaciones cara a cara - El método más eficiente y eficaz para transmitir información a y dentro de un equipo de desarrollo es la conversación cara a cara.
    Software de funcional - El software funcional es la principal medida del progreso.
    Trabajar a un ritmo sostenible Los procesos Agile promueven el desarrollo sostenible. Los patrocinadores, desarrolladores y usuarios deberían ser capaces de mantener un ritmo constante indefinidamente.
    Entorno Agile La atención continua a la excelencia técnica y el buen diseño mejora la agilidad.
    Simplicidad - El arte de maximizar la cantidad de trabajo no terminado, es esencial.
    Equipos autoorganizados: Las mejores arquitecturas, requisitos y diseños surgen de equipos autoorganizados.

<h1> Métodos Agile </h1>

Métodos Agile

El Manifiesto Agile por diseño, no era preciso sobre cómo debería funcionar Agile Después de forjar el Manifiesto, sus creadores (y muchos otros) siguieron evolucionando estas ideas, absorbiendo nuevas ideas de muchas fuentes y probándolas de manera práctica. Como resultado, en las últimas décadas, han surgido muchas versiones de Agile y algunas se han vuelto muy populares. Entre ellas se encuentran:

    Agile Scrum - En el rugby, el término "scrum" describe un punto en el juego donde los jugadores se juntan y tratan de ganar la posesión de la pelota. Scrum se enfoca en equipos pequeños y autoorganizados que se reúnen diariamente durante períodos cortos y trabajan en sprints iterativos, adaptando constantemente los entregables para cumplir con los requisitos cambiantes.
    Lean - Basado en Lean Manufacturing, el método Lean enfatiza la eliminación del esfuerzo desperdiciado en la planificación y ejecución y la reducción de la carga cognitiva del programador.
    Extreme Programming (XP): En comparación con Scrum, XP es más prescriptivo acerca de las mejores prácticas de ingeniería de software y aborda de manera más deliberada los tipos específicos de problemas de calidad de vida que enfrentan los equipos de desarrollo de software.
    Feature-Driven Development (FDD) FDD prescribe que el desarrollo de software debe proceder en términos de un modelo general, desglosado, planificado, diseñado y construido característica por característica. Especifica un proceso de modelado utilizado para estimar y planificar la entrega de funciones y un conjunto detallado de roles tanto para el equipo de desarrollo principal como para el personal de soporte.

De las metodologías descritas anteriormente, Agile Scrum es probablemente la más popular. A continuación analizaremos algunos términos y conceptos de Scrum que han sido adoptados más o menos universalmente por la comunidad Agile entre todas las metodologías.

Sprints

En el modelo Agile, el ciclo de vida del desarrollo de software aún se aplica. A diferencia del método Cascada (Waterfall), donde hay una iteración larga del SDLC, Agile es muchas iteraciones rápidas del SDLC.

Estas iteraciones rápidas se denominan sprints, y el propósito de los sprints es lograr la entrega frecuente del principio de software funcional del manifiesto Agile. Un sprint es un período de tiempo específico (encuadrado en el tiempo) que suele ser de entre dos y cuatro semanas, pero preferiblemente lo más breve posible. La duración del sprint debe determinarse antes de que comience el proceso y rara vez debería cambiar.

Durante un sprint, cada equipo asume tantas tareas, también conocidas como historias de usuario, como sientan que pueden realizar dentro de la duración del sprint. Cuando el sprint ha terminado, el software debe ser funcional y entregable, pero eso no significa necesariamente que se entregará; un sprint no siempre conduce a una versión, pero Agile requiere que el software siga siendo entregable.

Backlog (Lista de entregables)

Es la función del propietario del producto crear la lista priorizada de entregables (backlog) Este backlog se compone de todas las características del software, en una lista de prioridades. Las características del backlog son el resultado de la fase Requisitos y análisis, e incluyen características que no necesariamente estarán en la versión inmediata. Se pueden agregar nuevas funciones a la cartera de pedidos (Backlog) en cualquier momento, y el propietario del producto puede volver a priorizar la lista de trabajos pendientes en función de los comentarios de los clientes.

Historias de usuarios

Cuando una característica se acerca a la parte superior de la lista de prioridades, se divide en tareas más pequeñas llamadas historias de usuario. Cada historia de usuario debe ser lo suficientemente pequeña como para que un solo equipo pueda terminarla en un solo sprint. Si es demasiado grande para ser completado en un solo sprint, el equipo debería desglosarlo más. Debido a que el software debe ser entregable al final de cada sprint, la historia de usuario también debe cumplir con esas reglas.

Una historia de usuario es una simple declaración de lo que un usuario (o un rol) necesita y por qué lo necesita La plantilla sugerida para una historia de usuario es:

Como <user|role>, me gustaría<action>, por lo que<value|benefit>

Para completar una historia de usuario, es necesario completar todas las fases del SDLC. La historia del usuario en sí ya debería tener los requisitos definidos por el propietario del producto y el equipo que se encargará de la historia del usuario, debe idear un diseño para la tarea, implementarla y probarla.

Equipos de Scrum

Los equipos de Scrum son multifuncionales, colaborativos, autogestionados y autoempoderados. Idealmente, estos equipos de scrum no deberían tener más de 10 individuos, pero deberían ser lo suficientemente grandes como para terminar una historia de usuario dentro de un sprint.

Todos los días, cada equipo de scrum debe tener una reunión de discución o conforntación (Standup). Esta reunión de discución o confrontación(Standup) es una reunión que no debe durar más de 15 minutos y debe tener lugar a la misma hora todos los días. De hecho, se llama "standup" porque idealmente debería ser lo suficientemente corto como para que el equipo lo logre sin tener que sentarse.

El objetivo de esta reunión diaria es mantener a todos los miembros del equipo en sincronía con lo que cada persona ha logrado desde la última reunión, en lo que van a trabajar hasta la siguiente, y qué obstáculos tienen que podrían impedirles finalizar su tarea. El Scrum Master facilita estas reuniones (Standup) y su trabajo es reportar y/o ayudar a eliminar obstáculos.
    
<h1> Desarrollo de Software bajo metodología Lean </h1>
    
El desarrollo de software Lean (Lean Software Development) se basa en los principios de Lean Manufacturing, que se centran en minimizar los desperdicios y maximizar el valor para el cliente.

En su forma más simple, esta modalidad de desarrollo (Lean Software) ofrece solo lo que los clientes desean. En el libro Lean Software Development: Un kit de herramientas Agile, hay siete principios para lean:

    Eliminar desperdicios
    Amplificar el aprendizaje
    Decidir lo más tarde posible
    Entregar lo más rápido posible
    Empoderar al equipo
    Integridad en compilación
    Optimizar la totalidad

Eliminar desperdicios

El desperdicio es cualquier cosa que no agregue valor a los clientes. La definición del valor es subjetiva, sin embargo, es determinada por el cliente. La eliminación de desperdicios es el principio fundamental de Lean, del cual se derivan todos los demás principios.

Para eliminar dichos desperdicos, debe ser capaz de entender en que consisten. El desperdicio es cualquier cosa que no agregue valor directo al cliente. Hay siete desperdicios en desarrollo de software:

    Trabajo parcialmente hecho
    Procesos adicionales
    Características adicionales
    Cambio de tareas
    Espera
    Movimiento
    Defectos

Trabajo parcialmente hecho

El trabajo parcialmente hecho es un desperdicio porque:

    No añade ningún valor al cliente
    El tiempo y los recursos empleados en este trabajo podrían haberse utilizado en algo que es de valor para el cliente
    El trabajo generalmente no se mantiene, por lo que eventualmente se vuelve obsoleto

Procesos adicionales

Los procesos adicionales son como un montón de papeleo. Como resultado, son un desperdicio por casi las mismas razones que el trabajo parcialmente hecho.

Características adicionales

Si el cliente no lo pidió, no le aporta valor. Puede ser bueno tenerlo, pero es mejor usar los recursos para construir exactamente lo que quieren los clientes.

Cambio de tareas

Los humanos necesitan tiempo para cambiar su mente y centrarse en otra tarea, ese tiempo dedicado a cambiar contextos es un desperdicio. El cambio de tareas desperdicia el tiempo de un recurso (persona), por lo que es un desperdicio asignar un recurso a varios proyectos.

Espera

Mucha gente estaría de acuerdo en que, por definición, esperar es una gran pérdida de tiempo. Por lo tanto, cualquier tipo de retraso es un desperdicio. Ejemplos de un retraso en el desarrollo de software son los retrasos en:

    Iniciar el proyecto
    Obtener los recursos adecuados (personal)
    Obtener los requisitos del cliente
    Aprobaciones de la documentación
    Obtener respuestas
    Toma de decisiones
    Implementación
    Pruebas

Movimiento

El desarrollo de software Lean define el movimiento para dos cosas: personas y artefactos. El movimiento para las personas es cuando éstas necesitan caminar físicamente desde su escritorio a otro miembro del equipo para hacer una pregunta, colaborar, etc. Cuando se mueven desde su escritorio, no es sólo el tiempo que tardan en llegar al destino lo que es un desperdicio, sino también el tiempo para el cambio de tareas.

El movimiento de los artefactos es cuando los documentos o el código se mueven de una persona a otra. La mayoría de las veces, el documento no contiene toda la información que la siguiente persona necesita, por lo que esa persona tiene que reunir la información de nuevo, o la entrega requiere tiempo, lo que es un desperdicio.

Defectos

Los defectos desapercibidos (también conocidos como Bugs) son un desperdicio debido al impacto del defecto. El defecto puede causar un efecto de bola de nieve con otras características, por lo que el tiempo que tarda en depurarlo (Debuguing) es un desperdicio. Además, para un cliente, el valor del software se reduce cuando tiene problemas, por lo que la característica termina siendo un desperdicio.
    
<h1> Desarollo de Software Lean (Cont.) </h1>
    
Amplificar el aprendizaje con Sprints cortos

Para poder afinar el software, debe haber repeticiones cortas frecuentes de software funcional. Al tener más iteraciones:

    Los desarrolladores aprenden más rápido
    Los clientes pueden dar sus comentarios mas temprano
    Las funciones o características se pueden ajustar para que aporten más valor a los clientes

Decidir lo más tarde posible

Cuando hay incertidumbre, lo mejor es retrasar la toma de decisiones hasta lo más tarde posible en el proceso, porque es mejor basar las decisiones en hechos, en lugar de opiniones o especulaciones.

Además, cuando aún no se toma una decisión, el software está diseñado para ser flexible con el fin de adaptarse a la incertidumbre. Esta flexibilidad permite a los desarrolladores realizar cambios cuando se toma una decisión, o en el futuro, si los requisitos cambian.

Entregar lo más rápido posible

Entregar el software más rápido:

    Permite a los clientes proporcionar comentarios lo antes posible
    Permite a los desarrolladores ampliar el aprendizaje
    Ofrece a los clientes las características justo en el momento en que lo necesitan
    No permite a los clientes cambiar de opinión
    Hace que la toma de decisiones sea de una manera más rápida.
    Produce menos desperdicios

Notará que cada una de estas razones practica al menos uno de los principios Lean discutidos anteriormente.

Empoderar al equipo

Cada persona tiene su propia experiencia, así que déjelos tomar decisiones en su área de especialización. Cuando se combina con otros principios como eliminar desperdicos, tomar decisiones tardías y entregas rápidas, no hay tiempo para que otros tomen decisiones para el equipo.

Integridad en la compilación

La integridad del software es cuando el software aborda exactamente lo que el cliente necesita. Otro nivel de integridad es que el software mantiene su utilidad para el cliente.

Optimizar la totalidad

Aunque uno de los principios es empoderar al equipo, cada experto debe dar un paso atrás y ver el panorama general. El software debe construirse de forma coherente. El valor del software se verá afectado si cada experto se centra únicamente en su experiencia y no considera los efectos de sus decisiones sobre el resto del software.

<h1> Patrones de diseño de software </h1>

Introducción

Los patrones de diseño de software son soluciones de mejores prácticas para resolver problemas comunes en el desarrollo de software. Los patrones de diseño son independientes del lenguaje. Esto significa que pueden implementarse en cualquier lenguaje de programación contemporáneo, de uso general, o en cualquier lenguaje que admita programación orientada a objetos. A menudo, los patrones de diseño populares fomentan la creación de frameworks (Marco) complementarios que simplifican la implementación en lenguajes y paradigmas ampliamente utilizados.

Los artesanos siempre han compartido métodos y técnicas probados en el tiempo para la resolución de problemas. Llamar a estas cosas "patrones de diseño" se hizo por primera vez sistemáticamente en el campo de la arquitectura y la planificación urbana. Los patrones arquitectónicos se organizaron por clases abstractas de problemas resueltos, instando a los diseñadores a reconocer los temas centrales comunes compartidos en numerosos contextos divergentes. Por ejemplo, una parada de autobús y una sala de espera de hospital, son ambos lugares en los que la gente espera; por lo que ambos pueden implementar de manera útil características del patrón UN LUGAR PARA ESPERAR.

Esta forma de pensar sobre los patrones fue rápidamente recogido por los pioneros en la codificación orientada a objetos y el desarrollo de software Agile (Ágil). En 1994, Erich Gamma, Richard Helm, Ralph Johnson y John Vlissides (conocido colectivamente como la Pandilla de los Cuatro (GoF)) publicaron un libro llamado Design Patterns - Elements of Reusable Object-Oriented Software. Ofreceremos una visión amplia de los patrones que identificaron y documentaron.

Programa a una interfaz, no a una implementación. Acoplar estrechamente la lógica del programa principal con implementaciones de funcionalidad específica tiende a hacer que el código sea difícil de entender y mantener. La experiencia ha demostrado que funciona mejor acoplar capas lógicas sueltas mediante el uso de interfaces abstractas. Por ejemplo, el código de línea principal llama a funciones y métodos de una manera genérica. Las funciones de nivel inferior implementan interfaces de coincidencia para la funcionalidad que proporcionan, asegurando, por ejemplo, que todas las funciones de serialización utilizadas en un programa se llamen de manera similar.

Lenguajes orientados a objetos como Java formalizan estas ideas. Permiten la declaración explícita de interfaces que las clases pueden implementar. Una definición de interfaz es básicamente una colección de prototipos de funciones, definiendo nombres y tipos para funciones y parámetros que la lógica de nivel superior podría usar para invocar un rango de clases. Por ejemplo, la interfaz con un rango de clases de "vehículo" (por ejemplo, class "coche", "motocicleta", "tractor") podría incluir prototipos de arranque\ _motor (), parada\ _motor (), accelerar () y freno ().

Favorece la composición del objeto sobre la herencia de clase. Los lenguajes orientados a objetos permiten la herencia: clases base más generalizadas pueden ser heredadas por clases derivadas. Por lo tanto, una clase 'pato' podría heredar funcionalidad sustancial de una clase 'pájaro'. Esto requiere, sin embargo, que la clase de aves implemente una gama muy amplia de métodos, la mayoría de los cuales no pueden ser utilizados por una clase derivada específica.

El principio de favorecer la composición sobre la herencia sugiere que una mejor idea puede ser favorecer la implementación de una clase específica (class - pato) creando sólo subclases únicas requeridas (class - cuack) junto con interfaces abstractas (interfaz - 'comopato') a clases (class - volar, class - nadar) que pueden ser compartido ampliamente de manera similar (class - pingüino implementa interfaz 'comopingüino, 'que permite compartir la clase 'nadar', pero no' volar '). Organizar el software de esta manera ha demostrado ser más flexible, en última instancia más fácil de mantener, y fomenta la reutilización del código.

Los patrones de diseño de software ya han demostrado ser exitosos, por lo que su uso puede acelerar el desarrollo porque los desarrolladores no necesitan encontrar nuevas soluciones y pasar por una prueba de concepto para asegurarse de que funcionan.

<h1> Modelo-Vista-Controlador (MVC) </h1>

El patrón de diseño Modelo-Vista-Controlador (MVC) a veces se considera un patrón de diseño arquitectónico. Su objetivo es simplificar el desarrollo de aplicaciones que dependen de interfaces gráficas de usuario (GUI). MVC abstrae el código y la responsabilidad en tres componentes diferentes: modelo, vista y controlador. Cada componente se comunica entre sí en una dirección. Este patrón de diseño se utiliza comúnmente en interfaces de usuario y aplicaciones web.

Componentes

    Modelo - El modelo es la estructura de datos de la aplicación y es responsable de administrar los datos, la lógica y las reglas de la aplicación. Obtiene la entrada del controlador.
    Vista - La vista es la representación visual de los datos. Puede haber varias representaciones de los mismos datos.
    Controlador: El controlador es como el intermediario entre el modelo y la vista. Toma la entrada del usuario y la manipula para ajustarse al formato del modelo o la vista.

La ejecución del Modelo - Vista - Controlador luce de la siguiente manera

    El usuario proporciona entrada.
    El controlador acepta la entrada y manipula los datos.
    El controlador envía los datos manipulados al modelo.
    El modelo acepta los datos manipulados, los procesa y envía los datos seleccionados (en las formas más estrictas de MVC, a través del controlador) a la vista.
    La vista acepta los datos seleccionados y los muestra al usuario.
    El usuario ve los datos actualizados como resultado de su entrada.

La ventaja del patrón de diseño Modelo-Vista-Controlador es que cada componente se puede construir en paralelo. Debido a que cada componente se abstrae, la única información que necesita cada componente es la interfaz de entrada y salida para los otros dos componentes. Los componentes no necesitan saber acerca de la implementación dentro de los otros componentes. Además, dado que cada componente sólo depende de la entrada que recibe, los componentes se pueden reutilizar siempre y cuando los demás componentes proporcionen los datos de acuerdo con la interfaz correcta.

<h1> Codificación Básica </h1>
Métodos, funciones, módulos y clases

Puede parecer fácil escribir código en un archivo y hacerlo funcionar. Pero a medida que aumenta el tamaño y la complejidad del proyecto, y a medida que otros desarrolladores (y partes interesadas) se involucran, se necesitan métodos disciplinados y mejores prácticas para ayudar a los desarrolladores a escribir mejor código y colaborar en torno a él más fácilmente.

Una cosa en la que la mayoría de los desarrolladores están de acuerdo es intentar escribir código limpio. Pero, ¿qué es el código limpio?

<h2> Código limpio </h2>

El código limpio es el resultado cuando los desarrolladores intentan hacer que su código sea fácil de leer y entender para otros desarrolladores. Lo que constituye "código limpio" es algo subjetivo, pero aquí hay algunos principios comunes:

    ¿El formato del código es limpio, y sigue las prácticas de formato generalmente aceptadas para el o los lenguaje(s) de programación involucrados, y/o cumple con los requisitos específicos del "stylebook" institucional del proyecto o del equipo?
        ¿Se pega con TODAS las pestañas o TODOS los espacios?
        ¿Utiliza el mismo número de pestañas o espacios por nivel de la sangría en todo el archivo? (Algunos lenguajes, como Python, hacen de esto un requisito).
        ¿Tiene la sangría en los lugares correctos?
        ¿Utiliza un formato coherente para la sintaxis, como la ubicación de llaves ({})?
    ¿Se usan nombres variables, objetos y otros en el código intuitivos?
    ¿Está organizado el código de una manera que tenga sentido? Por ejemplo, ¿están agrupadas las declaraciones, con funciones en la parte superior, código principal en la parte inferior, o de otro modo, dependiendo del idioma y el contexto?
    ¿El código está documentado internamente con comentarios apropiados?
    ¿Cada línea de código sirve para un propósito? ¿Ha eliminado todo el código no utilizado?
    ¿Está escrito el código para que el código común se pueda reutilizar y para que todo el código se pueda probar fácilmente por unidad? 

El código limpio enfatiza:

    Formato estandarizado y nomenclatura intuitiva para facilitar la lectura, comprensión y búsqueda
    Organización general para comunicar la intención y hacer las cosas más fáciles de encontrar
    Modularidad para facilitar las pruebas y la reutilización
    Comentarios en línea y otros
    Otras características que ayudan a que el código se autodocumente

En teoría, otros desarrolladores deberían poder entender, usar y modificar su código sin necesidad de hacerle preguntas. Esto acelera enormemente el desarrollo, permitiendo la reutilización, depuración, análisis de seguridad, revisión de código y fusión, junto con otros procesos. Permite que los proyectos de larga data (por ejemplo, proyectos de código abierto) incorporen su código con mayor confianza. Y permite a las organizaciones seguir utilizando su código de manera eficiente, después de haber pasado a otros proyectos o trabajos.

Por el contrario, el código que no está limpio se convierte rápidamente en deuda técnica. Puede ser insostenible, o inutilizable con total confianza. Es un código que necesita ser refactorizado, o eliminado y reescrito, lo cual es costoso y requiere mucho tiempo.

¿Cuáles son algunas otras razones por las que los desarrolladores quieren escribir código limpio?

    El código limpio es más fácil de entender, más compacto y mejor organizado, lo que tiende a dar como resultado un código que funciona correctamente (menos errores) y funciona según sea necesario.
    El código limpio, al ser modular, tiende a ser más fácil de probar utilizando métodos automatizados como marcos de puebas unitarias.
    El código limpio, estandarizado, es más fácil de escanear y verificar utilizando herramientas automatizadas como linters, o herramientas de línea de comandos como grep, awk y sed.
    Simplemente se ve mejor.

Ahora que comprende los objetivos de escribir código limpio, puede profundizar en las mejores prácticas de codificación. Específicamente, mirando cómo dividir el código en métodos y funciones, módulos y clases.
