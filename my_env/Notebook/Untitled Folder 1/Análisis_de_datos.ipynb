<h3> Tipo de gráficas de mapas </h3>
<p>
Como comentamos en la clase pasada, las visualizaciones de datos en forma de mapas pueden tener un sin fin de representaciones pero lo importante es entender que nos referimos a una gráfica que representa una proporción de un área sobre una superficie plana.

Por lo general, estos mapas utilizan referencias de medición en los costados para poder entender la proporción en la cual fueron diseñados, por ejemplo: 1 cm = 1 km. Es importante recalcar que estas mediciones se utilizan para dar mayor fidelidad en la representación de información.

Entre los distintos tipos de representación en mapas que existe, podemos encontrar los más conocidos y utilizados:

Tipos de mapas: isolíneales, coropletas, diagramas, anamórficos.
</p>
<h3> Mapas isolíneales </h3>
<p>
Utilizan líneas para poder unir áreas que derminen cierta similitud relacionada con alguna variable. Estos espacios suelen colorearse para mostrar distinción entre ellos. Este tipo de mapas son muy utilizados en el ámbito meteoroógico, en el caso de la imagen podemos observar un mapa isobárico para medir la presión armosférica. 
</p>

<img src="https://static.platzi.com/media/user_upload/image2-be0a146f-bf82-4b8f-b05d-1729064a3602.jpg">

<h3> Mapas Coropléticos </h3>
<p>
Son mapas que representan con color los lugares donde ocurrieron ciertos sucesos(variable). Durante las elecciones presidenciales de los Estados Unidos, esos mapas se utilizaron para mostrar el partido político que ganó en determinados estados.
</p>
<img src="https://static.platzi.com/media/user_upload/image3-f63c325f-c63b-4bcc-a1b2-76f715350bf1.jpg">

<h3> Mapas de Diagramas </h3>
<p>
Estos mapas pueden considerarse como mapas combinados, ya que por lo general utilizan el color para hacer una distinción de variable y alguna otra visualización sobre cada segmento para mostrar una variable adicional. Generalmente estos mapas son utilizados en estadística geográfica, reportes económicos, de salud, etc.
</p>
<img src="https://static.platzi.com/media/user_upload/image1-653c8f87-70cb-4f8a-94e6-0a72d87754fa.jpg">

<h3> Gráfica de tablas </h3>

Data viz en forma de tablas tenemos la representación de forma ordenada en filas y columnas,  donde tenemos en las filas algun tipo de categorización y en las columnas lo que significa cada una de las propiedades en la que hacen la intercepción de nuestros datos. 

Qué hacer: 
Se utilizan en la industria del business inteligent y en data science, cuando queremos entregar un mensaje acompañado de otro data visualization, ya que es más facil entender de forma gráfica pero con las tablas podemos resolver las dudas.

Qué no hacer: 

No utilizar datos tan extensos, ya que la audencia no va entender el mensaje.
