<h4>Qué es postgresql?</h4>
es un sistema de bases de datos de objetos relacionales (ORDBMS-object-relational database management system ) 
desarrollada en el departamento de ciencas de la computanción en la universidad de Berkeley en California. POSTGRES es pionera de muchos conceptos que solo fueron displonibles mas tarde en algunos sistemas de base de datos comerciales.
    
Postgresql es de codigo-abierto decendiente de este códifo orininal de Berkeley. Esta soportada en gran parte por el estandar de SQL y ofrece muchas caracteristicas modernas: 
        
    Consultas complejas.
    llaves foraneas.
    Disipadores. 
    Vistas de actualizables. 
    Integridad transaccional.
    Control de concurrencia de múltiples versiones. 
    
Además, PostgreSQL puede ser extendida por el usuario de muchas maneras, por ejemplo agregando nuevos: 
    
    Tipos de datos.
    Funciones.
    Operadores.
    Funciones agregadas.
    Métodos indexados.
    Lenguajes de procedimiento.

y porque la licencia es libre, PostgreSQL puede ser, módificada, y distribuida por cualquer persona de forma gratuita para cualquier proposito, puede ser privada, comercial, o académico.
    
    
    
Concepto de lenguajes de datos: 

El motor de base de datos mas popular del mundo.

Lenguaje:  Lo que permite agregar información o realizar querys en una base de datos. Ejemplo SQL

Motor: Permite estructurar la información dentro de un servidor. Ejemplo: postgreSQL, MySQL.

Servidor: Un equipo el cual tiene una memoría y una ram para que funcione el motor de la base de datos.

Open Source: Un equipo el cual tiene una memoría y una ram para que funcione el motor de base de datos.

Objeto Relacional: Se programa teniendo como referencia también la programación orientada a objetos por temas como la herencia.

PostGIS: plugin para recibir información de sistemas información geográfica.

PL/PgSQL: Lenguaje backend interno del motor de base de datos.

Cumple ACID: Cantidad de reglas que deben cumplir las bases de datos para garantizar buenas practicas.
    
A: Atomicity: Puedo separar las funciones programadas para la base de datos en pequeñas tareas y desarrolladas como un todo de modo que cuando ejecutas todas las tareas pero si alguna de ellas falla se hace un rolback, o se deshacen los cambios de cada una de las que se habian pasado. 
    
C: Consistency: Todo lo que se desarrollo entorno al objeto relacional en el que tu puedes tener dos tablas que están relacionadas con llaves primarias, llaves foráneas.

I: Isolation: (Aislamiento) Puedes tener varias tareas ejecutandose al mismo tiempo 
    dentro de la base de datos, puedes tener consultas y al mismo tiempo tener modificaciones desde una tabla diferente,  puedes tener roles diferentes accediendo a la base de datos. Es decir puedes tener aisladas las funciones que se ejecutan dentro de la base de datos.
    
    
D: Durability: (Durabilidad) Puedes tener la seguridad de que la información no se va a perder. Digamos que el servidor donde estaba corriendo, se apagó durante una inserción, esa información no se va a perder por que postgreSQL uso un sustema en el guarda la información en una bitácora y luego hace el cambio efectivo en la base de datos.

Practica: 

Crear una base de datos: CREATE DATABASE transporte;
Crear la tabla tren: CREATE TABLE tren(id     serial NOT NULL,
                                       modelo character varying,
                                       capacidad integer,
                                       CONSTRAINT tren_pkey
                                       PRIMARY KEY (id)
                                      );

La columna id será un número autoincremental (cada vez que se inserta un registro se aumenta en uno), modelo se refiere a una referencia al tren, capacidad sería la cantidad de pasajeros que puede transportar y al final agregamos la llave primaria que será id.

Vemos la tabla usando el comando: \d tren.

Podemos ver en los campos de la tabla en la consola, la columna 'Default', clasifica a id como nextval('tren_id_seq'::regclass)

si ejecutamos \d tren_id_seq : 

    Podemos notar que es tipo entero, comienza en 1, el mínimo valor que toma es 1, el máximo valor que toma 
    es 2147483647, tiene  un incremento de 1. 
    
Investigar más sobre \d tren_id_seq.

Ingresemos un dato en nuestra tabla con el siguiente comando: 
        INSERT INTO tren
                    (modelo, capacidad) 
                    VALUES 
                    ('Volvo 1', 100);
                    
Consultemos los datos en la tabla: SELECT * FROM tren;

Vamos a modíficar el valor, establecer el tren con id uno que sea modelo Honda 0726. Para ello ejecutamos la consulta tipo actualización: 

UPDATE tren SET modelo = 'Honda 0726' WHERE id =1;

utilicemos el comando SELECT * FROM tree, para verificar que el elemento ha cambiado. 

Ahora borramos la fila con el comando: 

DELETE FROM tren WHERE id = 1;

Ahora verifiquemos que se ha eliminado la fila.

Para medir el tiempo en que se demora una consulta utilizamos el comando : 

      \timing
